```python
import spectral.io.envi as envi
from Model import KmeansSegmentation
import glob
import os
import numpy as np
import matplotlib.pyplot as plt

# ListOfDir= the path to your directory, where your data is, that has FOLDERs that contain .hdr and .dat extensions
# The segmented labels will be stored in the Segmented_Images folder as numpy files
ListOfDir=glob.glob(r'./HSI_Data/*')

for dir in ListOfDir:
    hsi_ref = envi.open(os.path.join(dir, dir[dir.rfind('\\')+1:]+'.hdr'),\
                      os.path.join(dir, dir[dir.rfind('\\')+1:]+'.dat'))
    hsi_np = hsi_ref.asarray()
    SegmentedImage=KmeansSegmentation.segment(hsi_np, show_figure=True)
    plt.title(dir[dir.rfind('\\')+1:])
    np.save(r'./Segmented_Images/'+str(dir[dir.rfind('\\')+1:]),SegmentedImage)
```
